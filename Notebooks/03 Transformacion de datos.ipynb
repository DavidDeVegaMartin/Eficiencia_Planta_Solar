## 3.1 Set-Up

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

%config IPCompleter.greedy = True

pd.options.display.float_format='{:15.2f}'.format
```

## 3.2 Carga de Datos

```python
# Carga Archivo
df = pd.read_csv('..//Datos/df_procesado.csv')

# Convertir la columna 'fecha' a tipo datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Establecerla como Ã­ndice
df.set_index('fecha', inplace=True)
df.head()
```

```python
df.info()
```

## 3.3 Generacion de variables

### 3.3.1 Dimension Tiempo

```python
def componentes_fecha(dataframe):
    mes = dataframe.index.month
    dia = dataframe.index.day
    hora = dataframe.index.hour
    minuto = dataframe.index.minute

    return pd.DataFrame({'mes':mes, 'dia':dia, 'hora':hora, 'minuto':minuto})
```

```python
df = pd.concat([df.reset_index() ,componentes_fecha(df)], axis = 1).set_index('fecha')
df
```

### 3.3.2 Eficiencia Inverters

```python
def eficiencia_inverter(AC,DC):
    temp = AC / DC * 100
    return(temp.fillna(0))
```

```python
df['eficiencia'] =  eficiencia_inverter(df.kw_ac, df.kw_dc)
```

```python
df.eficiencia.isna().sum()
```

```python
df.eficiencia.plot.kde()
plt.show()
```

```python
# Estilo estÃ©tico general
sns.set(style="whitegrid", palette="muted", font_scale=1.2)

# TamaÃ±o de la figura
plt.figure(figsize=(10, 6))

# KDE plot con color, grosor, sombra y suavizado
sns.kdeplot(data=df, x='eficiencia', fill=True, color="#4C72B0", linewidth=2.5, alpha=0.6)

# TÃ­tulo elegante y etiquetas
plt.title("DistribuciÃ³n de la Eficiencia AC-DC", fontsize=18, fontweight='bold', color="#333333")
plt.xlabel("Eficiencia", fontsize=14)
plt.ylabel("Densidad", fontsize=14)

# Personalizar fondo y ejes
plt.gca().set_facecolor("#F5F5F5")
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Mostrar grÃ¡fico
plt.tight_layout()
plt.show()
```

## 3.4 Reordenacion del dataframe

```python
# Reordenamos el dataframe para facilitar la visualizacion del mismo.
orden = ['planta','mes','dia','hora','minuto','sensor_id','irradiacion','t_ambiente','t_modulo',
         'inverter_id','kw_dc','kw_ac','eficiencia','kw_dia','kw_total']
df = df[orden]
df
```

Para facilitar la labor de analisis y visualizacion, cambiamos la temporalidad del dataframe a diario.

Esto comprende las siguientes tareas:

    1. Agregamos por Planta e Inverter (groupby) y transformaremos fecha mediante resample.

    2. Aplanar los multindices tanto en columnas como en filas.

1. Agregamos por Planta e Inverter (groupby) y transformaremos fecha mediante resample.

## 3.5 CreaciÃ³n dataframe diario

```python
df_dia = df.groupby(['planta', 'inverter_id']).resample('D').agg({
    'irradiacion': ['min', 'mean', 'max'],
    't_ambiente': ['min', 'mean', 'max'],
    't_modulo': ['min', 'mean', 'max'],
    'kw_dc': ['min', 'mean', 'max','sum'],
    'kw_ac': ['min', 'mean', 'max','sum'],
    'eficiencia': ['min', 'mean', 'max'],
    'kw_dia': ['max'],
    'kw_total': ['max']})
```

```python
df_dia.head()
```

2.1 Aplanar los multindices tanto en columnas.

```python
# Eliminamos el multindex para facilitar el uso (to_flat_index)
tuplas = df_dia.columns.to_flat_index()
tuplas
```

```python
df_dia.columns = ['_'.join(par) for par in tuplas]
df_dia
```

2.2 Aplanar los multindices  en filas.

```python
df_dia = df_dia.reset_index().set_index('fecha')
df_dia
```
